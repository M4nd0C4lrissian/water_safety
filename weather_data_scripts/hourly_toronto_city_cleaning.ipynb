From inspection, seems to be missing all precipitation readins up until and not including 2014.

These need to be removed, and we'll check for more null values, of which Humidex has a lot - I need to see what a baseline humidex value might be.

I filled humidex values with a normal dist of them using the mean and std of the filled-in data. This is probably a bad idea - I just feel like humidity is super relevant for bacterial growth.

```python
df = pd.read_csv('cleaned_data\hourly\\merged_toronto_hanlans.csv', index_col=0)
```

```python
df.drop(columns=['Date/Time (LST)', 'Month', 'Year', 'Day'], inplace=True)
```

```python
mask = (df['Precip. Amount (mm)'].isna())
df.drop(index=df.loc[mask].index, inplace=True)
```

```python
df.describe()
```

```python
rng = np.random.default_rng(0)
mu = df['Hmdx'].mean()
sd = df['Hmdx'].std()

filler = pd.Series(rng.normal(loc=mu, scale=sd, size=df.shape[0]))
print(filler)
df['Hmdx'] = df['Hmdx'].fillna(filler)
```
